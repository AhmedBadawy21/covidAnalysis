```python
response = requests.get('https://gorest.co.in/public-api/users?_format=json&access-token=H4lYmIGLXZkDlpy_w0ZVS5rgcpDXbN7fhRRy')
employees = requests.get('https://gorest.co.in/public-api/users?_format=json&access-token=H4lYmIGLXZkDlpy_w0ZVS5rgcpDXbN7fhRRy')

# response is a module
response.text # gives u a tring code

employees = response.json()
employees["data"]
for e in employees["data"]:
    print(e['id'],'|',e['name'],'|',e['email'],'|',e['gender'],'|',1 if e['status']== 'active' else 0)
```

```python
with open('employees.csv','w',newline = '') as f: #newline = '' to stop making a new line while starting writing
    writer = csv.DictWriter(f, fieldnames=['id','name','email','gender','active'])
    writer.writeheader()
    for e in employees['data']:
        Xid = e['id']
        Xname = e['name']
        Xemail = e['email']
        Xgender = e['gender']
        Xactive = True if e['status'] == 'active' else False
        writer.writerow(
            {
                'id' : Xid,
                'name' : Xname,
                'email' : Xemail,
                'gender' : Xgender,
                'active' : Xactive
            }
        )
```

```python
response2 = requests.get('http://coronavirus.m.pipedream.net/')

covidData = response2.json()

with open('CovisAnalysis.csv','w',newline = '') as f: #newline = '' to stop making a new line while starting writing
    writer = csv.DictWriter(f, fieldnames=["Country_Region","Province_State","Confirmed","Deaths"])
    writer.writeheader()
    for e in covidData["rawData"]:
        country = e["Country_Region"]
        provinance = e["Province_State"]
        confirmed = e["Confirmed"]
        death = e["Deaths"]
        
        writer.writerow(
            {
                "Country_Region" : country,
                "Province_State" : provinance,
                "Confirmed" : confirmed,
                "Deaths" : death,
             
            }
        )
```
